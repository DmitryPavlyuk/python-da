```python
import csv

movies = {}
with open('movielens/movies.csv', 'r', encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for line in reader:
        movies[line['movieId']] = line['title']
        
movies
```

```python
import numpy as np
ratings = []
users = set()
with open('movielens/ratings.csv', 'r', encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for line in reader:
        users.add(line["userId"])
        ratings.append(line)
users = list(users)
movieIds = list(movies.keys())
rating_mat = np.zeros((len(users),len(movies)))
for rating in ratings:
    rating_mat[users.index(rating["userId"]),movieIds.index(rating["movieId"])] = float(rating["rating"])
```

```python
rating_mat
```

```python
user1_movies = rating_mat[:1]
for i in np.where(user1_movies>0)[1]:
    print(movies[movieIds[i]]+" - "+str(user1_movies[0,i]))
```

```python
U, s, V=np.linalg.svd(rating_mat,full_matrices=False)
s=np.diag(s)
```

```python
Usk=np.dot(U,np.sqrt(s))
skV=np.dot(np.sqrt(s),V)
UsV = np.dot(Usk, skV)
```

```python
user1_rec = UsV[:1]
for i in np.where(user1_rec>0)[1]:
    print(movies[movieIds[i]]+" - "+str(np.round(user1_rec[0,i])))
```
