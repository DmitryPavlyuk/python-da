```python
import cv2
```

## Loading video into NumPy array

```python
vid = cv2.VideoCapture('Cheetah30sec.mp4')
frames = []
check = True
while check:
    check, frame = vid.read()
    if frame is not None:
        frames.append(frame)
videoarr = np.array(frames,dtype=np.uint8)
```

Resulting 4D array's shape: [frames, height, width, colors]


```python
videoarr.shape
```

## Last frame

```python
last_frame = videoarr[-1,:,:,:]
Image.fromarray(last_frame)
```

## Last frame: correct colours

```python
Image.fromarray(last_frame[:,:,[2,1,0]])
```

## Processing video

Reverting the frames and changing the colout scheme

```python
import imageio
speed = 5
imageio.mimwrite('Cheetah30sec_back.mp4', videoarr[:0:-speed,:,:,[2,1,0]] , fps = 24*speed)
```
